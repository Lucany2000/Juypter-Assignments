### 6. Wrangling and Cleansing Data
---
*(Cleanse the data, fill up missing data, and wrangle the data to prepare for analysis. Explain the process and the results.)*

We started cleaning the data by putting 0 for any empty space. Then, we checked for duplicate rows based on the 'Player' column, and combine player statistics for players listed twice.

```python
# Check for missing values in the entire DataFrame
missing_values = NBA.isnull().sum()

# Check if there are any missing values (sum > 0)
if missing_values.sum() == 0:
    print("There are no missing values in the DataFrame.")
else:
    print("There are missing values in the DataFrame.")
    print(missing_values)
```

```python
# Fill missing values with 0
NBA.fillna(0, inplace=True)
```

```python
NBA_combined = NBA.groupby('Player').agg({
    'Position': 'first',
    'Age': 'first',
    'Team': 'first',
    'Games': 'sum',
    'Starter': 'sum',
    'Minutes': 'mean',
    'Shots Made': 'sum',
    'Shot Attempts': 'sum',
    'Percentage Made': 'mean',
    '3 pointers': 'sum',
    '3 attempts': 'sum',
    '3 percentage': 'mean',
    '2 pointer': 'sum',
    '2 attempts': 'sum',
    '2 percentage': 'mean',
    'Efficient shot percentage': 'mean',
    'Free Throws': 'sum',
    'FT Attempts': 'sum',
    'FT %': 'mean',
    'Off Rebounds': 'sum',
    'Def Rebonds': 'sum',
    'Total Rebounds': 'sum',
    'Assists': 'sum',
    'Steals': 'sum',
    'Blocks': 'sum',
    'Turnovers': 'sum',
    'Fouls': 'sum',
    'Total Points': 'sum'
}).reset_index()
```

```python
# Check for duplicates based on the 'Player' column
duplicate_players = NBA[NBA.duplicated(subset='Player', keep=False)]

# Display the duplicate rows based on the 'Player' column
print("Duplicate Rows Based on 'Player' Column:")
print(duplicate_players)
```

Now, we analyzed the shot percentage by using columns Percentage Made, 3 percentage, 2 percentage, and Efficient shot percentage based on the cleaned data. We also created a graph to look at the frequency of shots overall in each category, not based on specific players.

```python
# Assuming NBA_combined is your DataFrame
selected_columns = ['Percentage Made', '3 percentage', '2 percentage', 'Efficient shot percentage']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
fig.suptitle('Comparison of Shooting Percentages')

# Flatten the 2D array of subplots for easy indexing
axes = axes.flatten()

# Iterate through the selected columns and plot each one
for i, column in enumerate(selected_columns):
    ax = axes[i]
    ax.hist(NBA_combined[column], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(column)
    ax.set_xlabel('Percentage')
    ax.set_ylabel('Frequency')

# Adjust layout to prevent overlapping titles
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Show the plots
plt.show()
```

#Avg Play Time Position Stats

Most players in each position are getting about 11 games which is more than 1/2 of the total games played so far in the season. About 1/4 of the average are starters and each player plays a little less than 1/2 the full game.

```python


# Calculate the average age, games played, games started, and time played per position
position_stats = NBA.groupby('Position').agg({
    'Age': 'mean',
    'Games': 'mean',
    'Starter': 'mean',
    'Minutes': 'mean'
}).reset_index()

# Round the float columns to 2 decimal places
position_stats = position_stats.round({'Age': 2, 'Games': 2, 'Starter': 2, 'Minutes': 2})

# Sort the table in reverse order based on 'Pos'
position_stats = position_stats.sort_values(by='Position', ascending=False)

# Display the table with reduced floats
position_stats
```
